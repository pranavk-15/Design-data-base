```python
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

```python
import pandas as pd, numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
eco= pd.read_csv('ECO.csv')
eco.head()
```

```python
eco.info()
```

```python
eco.isnull().sum()
```

## Checking Duplicate value on ECO number attribute column

```python
eco[eco.duplicated('ECO number')]
```

### Droping duplicate ECO number

```python
eco = eco.drop_duplicates('ECO number')
eco
```

```python
eco.isnull().sum()
```

```python
#checking null value for Report filled
eco[eco['Reports filled'].isnull()]
```

```python
#imputing null value of Report filled with Nill.

eco['Reports filled'] = eco['Reports filled'].fillna(eco['Reports filled']=='Nill' )
eco['Reports filled'].isnull().sum()
```

```python
# As there is no Mech drawing so computing it with 0
eco['Mech'] = eco['Mech'].fillna(eco['Mech']==0 )
#checking null value in Mech.
eco['Mech'].isnull().sum()
```

```python
eco.Mech = eco.Mech.astype('int32')
eco.head()
```

```python
# As there is no Electrical drawing so computing it with 0
eco['Electrical'] = eco['Electrical'].fillna(eco['Electrical']== 0 )
#checking null value in Mech.
eco['Electrical'].isnull().sum()
```

```python
eco.Electrical = eco.Electrical.astype('int32')
eco.head()
```

```python
# As there is no Casting drawing so computing it with 0
eco['Casting'] = eco['Casting'].fillna(eco['Casting']==0 )
#checking null value in Mech.
eco['Casting'].isnull().sum()
```

```python
eco.Casting = eco.Casting.astype('int32')
eco.head()
```

```python
#Merging status and Review to new column Action
eco['Action'] = eco['Status']+' '+eco['Remark']
eco.head()
```

### Droping status and Review column

```python
eco.drop('Status',axis = 1 , inplace= True)
```

```python
eco.drop('Remark',axis = 1 , inplace= True)
```

```python
eco.head()
```

```python
eco['Red dwgs count'] = eco['Red dwgs count'].fillna(eco['Red dwgs count']==0 )
eco['Red dwgs count'].isnull().sum()
```

```python
eco['Red dwgs count'] = eco['Red dwgs count'].astype('int32')
eco.head()
```

```python
eco['Green dwgs count'] = eco['Green dwgs count'].fillna(eco['Green dwgs count']==0 )
eco['Green dwgs count'].isnull().sum()
```

```python
eco['Green dwgs count'] = eco['Green dwgs count'].astype('int32')
eco.head()
```

```python
eco['Time Ganesh'] = eco['Time Ganesh'].fillna(eco['Time Ganesh']==0 )
eco['Time Ganesh'].isnull().sum()
```

```python
eco['Time Ganesh'] = eco['Time Ganesh'].astype('float64')
eco.head()
```

```python
eco['Time Pranav'] = eco['Time Pranav'].fillna(eco['Time Pranav']==0 )
eco['Time Pranav'].isnull().sum()
```

```python
eco['Time Pranav'] = eco['Time Pranav'].astype('float')
eco.head()
```

```python
eco['Time Rohan'] = eco['Time Rohan'].fillna(eco['Time Rohan']==0 )
eco['Time Rohan'].isnull().sum()
```

```python
eco['Time Rohan'] = eco['Time Rohan'].astype('float64')
eco.head()
```

```python
eco['Time Payal'] = eco['Time Payal'].fillna(eco['Time Payal']==0 )
eco['Time Payal'].isnull().sum()
```

```python
eco= eco[~(eco['Time Payal']=='\\')]
```

```python
eco= eco[~(eco['Time Payal']=='Leave')]
```

```python
eco.head()
```

```python
eco['Time Payal'] = eco['Time Payal'].astype('float64')
eco.head()
```

```python
#for Time Dhanraj there is lot of value is missing so droping the column
eco.drop('Time Dhanraj',axis = 1 , inplace= True)
```

```python
eco.head()
```

```python
eco.isnull().sum()
```

```python
eco[eco['Action taken'].isnull()]
```

```python
# droping null value for action column
eco = eco[~(eco['Action taken'].isnull())]
```

```python
eco.isnull().sum()
```

```python
eco[eco['Action'].isnull()]
```

```python
eco = eco[~(eco['Action'].isnull())]
```

```python
eco.isnull().sum()
```

## Now creating Year month and date column 

```python
eco["Year"] = pd.DatetimeIndex(eco["Date"]).year
eco.head()
```

```python
Drawing_cat = np.array(['Mech','Electrical','Casting'])
Total = np.array([eco.Mech.sum(), eco.Electrical.sum(), eco.Casting.sum()] )
```

```python
plt.bar(Drawing_cat,Total,width = 0.5,align= 'center',edgecolor = 'black',color ='green' )
plt.title("Drawing count Acc. Category\n", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Black'})
# Labeling Axes
plt.xlabel("Total Count", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Black'})
plt.ylabel("Category", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Black'})

plt.show()

# from below we can observe Mechanical drawing check is highest and casting is too low
```
